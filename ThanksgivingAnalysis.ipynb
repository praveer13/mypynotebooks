```python
#datatype is object
```

```python
def sAgeInt(x):
    if pandas.isnull(x):
        return None
    age_int = x.split(' ')[0]
    age_int = age_int.replace('+','')
    return int(age_int)

age_int_col = data['Age'].apply(sAgeInt)
data['int_age'] = age_int_col    
```

```python
data['int_age'].describe()
```

```python
#pandas automatically converts int to float if there are NaN values
def sIncInt(x):
    if pandas.isnull(x):
        return None
    int_inc = x.split(' ')[0]
    if int_inc == 'Prefer':
        return None
    int_inc = int_inc.replace('$','')
    int_inc = int_inc.replace(',','')
    return int(int_inc)

data['int_income'] = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(sIncInt)

data['int_income'].describe()
```

```python
data_und50k = data.loc[data['int_income'] < 50000]
```

```python
data_und50k
```

```python
data_und50k['How far will you travel for Thanksgiving?'].value_counts()
```

```python
data_over150k = data.loc[data['int_income'] > 150000]
```

```python
data_over150k['How far will you travel for Thanksgiving?'].value_counts()
```
